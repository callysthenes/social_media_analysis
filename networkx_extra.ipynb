```python
import networkx as nx
import scipy
from networkx import *
import pandas as pd
```

```python
# Create a graph
G = nx.Graph()

# Create a graph with 7 vertices, one isolated vertix, and directed edges between them
G.add_nodes_from([1,2,3,4,5,6,7])
G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(1,4)
G.add_edge(2,3)
G.add_edge(2,4)
G.add_edge(3,4)
G.add_edge(3,5)
G.add_edge(3,6)
G.add_edge(4,5)
G.add_edge(4,6)
G.add_edge(5,6)
G.add_edge(5,7)
G.add_edge(6,7)
G.add_edge(7,1)
G.add_edge(7,4)

# Draw the graph
nx.draw(G, with_labels=True)
```

```python

# Print the adjacency matrix
print(nx.adjacency_matrix(G).todense())
```

```python
#df of degree centrality and betweenness centrality
df = pd.DataFrame() 
df['degree_centrality'] = nx.degree_centrality(G)
df['betweenness_centrality'] = nx.betweenness_centrality(G)
 
#append df of eigenvector centrality
df['eigenvector_centrality'] = nx.eigenvector_centrality(G)
  
#append a df of closeness centrality
df['closeness_centrality'] = nx.closeness_centrality(G)
 
#append a df of pagerank
df['pagerank'] = nx.pagerank(G)



# Print the centrality measures
print(df)
 
```

```python

# Print the degree matrix in a tabular format 
print(nx.degree_centrality(G)
```

```python

# Print the laplacian matrix
print(nx.laplacian_matrix(G).todense())

# Print the normalized laplacian matrix
print(nx.normalized_laplacian_matrix(G).todense())

# Print the modularity matrix
print(nx.modularity_matrix(G).todense())
 
 
```
